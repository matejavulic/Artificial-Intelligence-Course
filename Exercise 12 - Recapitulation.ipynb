```python
#3.2 Napraviti dataset od dve koncentricne kruznice. Pronaci nov
#parametar uz pomoc kog je moguce izvrsiti LDA klasifikaciju
#ovakvog skupa podataka. Podeliti na trening/test skup u odnosu
#80 prema 20%.
```

```python
r_eda = np.random.rand(1000) * 5
ugao_eda = np.random.rand(1000)*np.pi*2

x1_eda = r_eda * np.cos(ugao_eda) 
x2_eda = r_eda * np.sin(ugao_eda)

ind1_eda = (r_eda<2) 
ind2_eda = (r_eda>4)

plt.scatter(x1_eda[ind1_eda],x2_eda[ind1_eda]) 
plt.scatter(x1_eda[ind2_eda],x2_eda[ind2_eda],c = 'r') 
plt.axis('equal')
plt.show()
```

```python
X4_1 = np.vstack([x1_eda[ind1_eda],[x2_eda[ind1_eda]]]).T
X4_2 = np.vstack([x1_eda[ind2_eda],[x2_eda[ind2_eda]]]).T

X4 = np.vstack([X4_1,X4_2])
y4 = np.arange(len(X4))
y4[:410] = 0
y4[410:] = 1
```

```python
X4_1.shape
```

```python
X4_2.shape
```

```python
#Promesajmo sve i podelimo na trening i test skup
```

```python
ind_sve = np.arange(len(y4))
np.random.shuffle(ind_sve)
```

```python
ind_train = ind_sve[:int(0.8*len(ind_sve))]
ind_test = ind_sve[int(0.8*len(ind_sve)):]
```

```python
b1 = np.hstack([X4_1[:,0],X4_2[:,0]])
b2 = np.hstack([X4_1[:,1],X4_2[:,1]])
```

```python
XX = np.sqrt(b1**2+b2**2).reshape(-1,1) #racunamo udaljenost od centra (novi parametar)
```

```python
XX.shape
```

```python
XX_train = XX[ind_train]
XX_test = XX[ind_test]
y_train = y4[ind_train]
y_test = y4[ind_test]
```

```python
#Sada uvozimo LDA
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
LDA = LinearDiscriminantAnalysis()
```

```python
LDA.fit(XX_train,y_train)
```

```python
pred = LDA.predict(XX_test)
```

```python
sum(pred == y_test)/len(y_test)*100 #--> Python3
```

```python
float(sum(pred == y_test))/float(len( y_test))*100 #--> Python2
```

# 5. DEO - Gradient Descent

```python
#5.1 Koristeci algoritam Gradient Descent-a odrediti minimum funkcije f(x) = 3(x1+3)^2 + 3(x2-6)^2 - 1
```

```python
w0 = np.linspace(-10,10,100)
w1 = np.linspace(-10,10,100)
f = 3*(w0+3)**2 + 3*(w1-6)**2 - 1
```

```python
import sys

from matplotlib.ticker import MaxNLocator
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

from scipy import linspace, meshgrid, arange, empty, concatenate, newaxis, shape
```

```python
x = sorted(w0)
y = sorted(w1)

X, Y = meshgrid(x, y)
Z = 3*(X+3)**2 + 3*(Y-6)**2 - 1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet, linewidth=0)
fig.colorbar(surf)

ax.xaxis.set_major_locator(MaxNLocator(5))
ax.yaxis.set_major_locator(MaxNLocator(6))
ax.zaxis.set_major_locator(MaxNLocator(5))

fig.tight_layout()
plt.show()
```

```python
plt.plot(w0, f)
plt.show()
```

```python
plt.plot(w1, f)
plt.show()
```

```python
iteracija = 1000
alfa = 0.1

w0 = np.random.randn(1)
w1 = np.random.randn(1)

w0_niz = []
w1_niz = []
f_niz = []

for i in range(iteracija):
    
    dw0 = 6*w0 + 18
    dw1 = 6*w1 - 36
    
    w0 = w0 - alfa*dw0
    w1 = w1 - alfa*dw1
    
    w0_niz.append(w0)
    w1_niz.append(w1)
    
    f = 3*(w0+3)**2 + 3*(w1-6)**2 - 1
    f_niz.append(f)
```

```python
print("Funkcija ima minimum za x1 =",float(w0),"i x2 =",float(w1))
```
